# Branches

always we see a `(master)` branch that changes start in it. but when we need to work on something else simultaneously, we should create new branch (for example writing Help or Two-language UI) and after `sub-branch` finished, then merge into `master` branch finally.

another application of branches, is mutiple working on a project at the same time. actually we never make changes in `master` branch, and each member of team work on a specific job on a `sub-branch`.

first command is:

    $ git branch

# Merge

for merge the `sub-branch` we did creat to `master` branch by this commands:

    $ git checkout master
    $ git merge fixpages

this command merge `fixpages` into `master` branch.

# Remotes

two phrases `push` and `pull` uses for sending files to `Network` and get files from it respectfully.

the `clone` phrase also use for setting a link to be main directory or `repository` for saving files. this link could be from `github` site or from a local srver as `gitlab` for example.

# PUSH and PULL

if we change sth on `local` and want to send to `Network` use this command:

    $ git push master origin

this command pushes master files into origin location on internet.

for pulling sown what is on `Network` in `master` directory in computer use this:

    $ git pull origin master
